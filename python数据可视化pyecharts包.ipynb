# 1、柱状图

```python
#1、导入相关图包---如from pyecharts.charts import Bar
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.globals import ThemeType

#2、设置画图数据(行名和数据)---如columns=[], data1=[], data2=[]
columns =["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
data1=[5, 20, 36, 10, 75, 90]

#3、用add()方法画图和进行图表设置
bar = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))     #使用主题
    .add_xaxis(columns)
    .add_yaxis("标签1", data1)
    .set_global_opts(title_opts=opts.TitleOpts(title="主标题", subtitle="副标题"))
)

#4、用render()方法来进行图表保存
bar.render("bar.html") #生成本地html文件
bar.render_notebook()  #Jupyter Notebook 直接调用 render_notebook 随时随地渲染图表
```

# 2、双柱图

```python
#1、导入相关图包---如from pyecharts.charts import Bar
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.globals import ThemeType

#2、设置画图数据(行名和数据)---如columns=[], data1=[], data2=[]
columns =["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
data1=[5, 20, 36, 10, 75, 90]
data2=[15, 6, 45, 20, 35, 66]

#3、用add()方法画图和进行图表设置
bar = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))     #使用主题
    .add_xaxis(columns)
    .add_yaxis("商家1", data1)
    .add_yaxis("商家2", data2)
    .set_global_opts(title_opts=opts.TitleOpts(title="主标题", subtitle="副标题"))
)

#4、用render()方法来进行图表保存
bar.render("bar2.html") #生成本地html文件
bar.render_notebook()  #Jupyter Notebook 直接调用 render_notebook 随时随地渲染图表
```

# 3、仪表盘

```python
#1、导入相关图包---如from pyecharts.charts import Bar
from pyecharts.charts import Gauge
from pyecharts import options as opts
from pyecharts.globals import ThemeType

#2、设置画图数据(行名和数据)---如columns=[], data1=[], data2=[]   

#3、用add()方法画图和进行图表设置
gauge= (
    Gauge()
    .add("", [("完成率", 66.6)])
    .set_global_opts(title_opts=opts.TitleOpts(title="Gauge-基本示例"))
)

#4、用render()方法来进行图表保存
gauge.render("gauge.html") #生成本地html文件
gauge.render_notebook()  #Jupyter Notebook 直接调用 render_notebook 随时随地渲染图表
```

# 5、词云

```python
#1、导入相关图包---如from pyecharts.charts import Bar
from pyecharts.charts import Page, WordCloud
from pyecharts.globals import SymbolType
from pyecharts import options as opts
from pyecharts.globals import ThemeType

#2、设置画图数据(行名和数据)---如columns=[], data1=[], data2=[]   
words = [
    ("Sam S Club", 10000),
    ("Macys", 6181),
    ("Amy Schumer", 4386),
    ("Jurassic World", 4055),
    ("Charter Communications", 2467),
    ("Chick Fil A", 2244),
    ("Planet Fitness", 1868),
    ("Pitch Perfect", 1484),
    ("Express", 1112),
    ("Home", 865),
    ("Johnny Depp", 847),
    ("Lena Dunham", 582),
    ("Lewis Hamilton", 555),
    ("KXAN", 550),
    ("Mary Ellen Mark", 462),
    ("Farrah Abraham", 366),
    ("Rita Ora", 360),
    ("Serena Williams", 282),
    ("NCAA baseball tournament", 273),
    ("Point Break", 265),
]

#3、用add()方法画图和进行图表设置
word = (
        WordCloud()
        .add("", words, word_size_range=[20, 100])
        .set_global_opts(title_opts=opts.TitleOpts(title="WordCloud-基本示例"))
    )

#4、用render()方法来进行图表保存
word.render("word.html") #生成本地html文件
word.render_notebook()  #Jupyter Notebook 直接调用 render_notebook 随时随地渲染图表
```
