# 建立坐标系

下面将讲解几种建立坐标系的方法，第一种add_subplot属于对象式编程，所有的对象都是针对某个对象进行；后三种事函数式编程，直接调用plt库里的函数，绘图语句必须马上跟在后面。对象式编程代码比较繁琐，但是便于理解；函数式编程虽然代码简洁，但是不容易掌握绘图原理。

1、用add_subplot函数建立坐标系

```python
#利用该函数建立坐标系时需要先建立画布
fig=plt.figure() #这段代码之前已经声明了也要声明，发现每次画都要声明
ax=fig.add_subplot(1,1,1)  #绘制1*1个坐标系
```

```python
fig=plt.figure() #这段代码之前已经声明了也要声明，发现每次画都要声明
ax1=fig.add_subplot(2,2,1)  #绘制2*2个坐标系，即4个坐标系，将位置1赋值给ax1
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
```

2、用plt.subplot2grid函数建立坐标系

```python
#该函数建立坐标系时不需要先建立画布，具体的绘图代码要跟在建立坐标系的语句后面
plt.subplot2grid((2,2),(0,0))  #表示将图表分成2行2列，且在（0，0）位置绘图
```

```python
import numpy as np
x=np.arange(6)
y=np.arange(6)
#将区域分成2行2列，且在位置(0,0)位置做折线图
plt.subplot2grid((2,2),(0,0)) 
plt.plot(x,y)
#将区域分成2行2列，且在位置(0,1)位置做柱形图
plt.subplot2grid((2,2),(0,1)) 
plt.bar(x,y)
```

3、用plt.subplot函数建立坐标系

```python
#该函数建立坐标系时不需要先建立画布，具体的绘图代码要跟在建立坐标系的语句后面
plt.subplot(2,2,1)  #表示将图表分成2行2列，且在第1个坐标系里绘图
```

```python
import numpy as np
x=np.arange(6)
y=np.arange(6)
plt.subplot(2,2,1)  #表示将图表分成2行2列，且在第1个坐标系里绘图
plt.plot(x,y)
plt.subplot(2,2,4)  #表示将图表分成2行2列，且在第4个坐标系里绘图
plt.bar(x,y)
```

4、用plt.subplots函数建立坐标系

plt.subplot2grid和plt.subplot每次只能返回一个坐标系，而plt.subplots能一次返回多个坐标系；该函数建立坐标系时不需要先建立画布，具体的绘图代码要跟在建立坐标系的语句后面

```python
fig,axes=plt.subplots(2,2) #绘制2*2个坐标系，即4个坐标系
```

```python
import numpy as np
x=np.arange(6)
y=np.arange(6)
fig,axes=plt.subplots(2,2) 
#表示在[0,0]坐标系中绘制折线图
axes[0,0].plot(x,y)
#表示在[1,0]坐标系中绘制柱形图
axes[1,1].bar(x,y)
```

1、坐标轴标签---plt.xlabel(labels)函数---labels表示标签值

```python
plt.xlabel('month')
plt.ylabel('amount')
x=np.arange(6)
y=np.arange(6)
plt.plot(x,y)
```

```python
#还可以设置标签距轴的距离-labelpad参数，设置字体的大小-xx-large参数，设置字体颜色-color参数,设置是否加粗-fontweight参数
plt.xlabel('month',labelpad=10,fontsize='x-large',color='#70AD47',fontweight='bold')
plt.ylabel('amount')
x=np.arange(6)
y=np.arange(6)
plt.plot(x,y)
```

2、坐标轴刻度---plt.xticks(ticks,labels)函数---tickes表示刻度值，labels表示标签值；---官方文档：https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlabel.html

```python
plt.xlabel('month')
plt.ylabel('amount')
plt.xticks(np.arange(6),['m1','m2','m3','m4','m5','m6'])
plt.yticks(np.arange(100,700,100),[100,200,300,400,500,600])
x=np.arange(6)
y=np.arange(100,700,100)
plt.plot(x,y)
```

```python
#有时候为里数据安全，不方便把坐标轴数值显示出来，这时传入一个空列表即可
plt.xlabel('month')
plt.ylabel('amount')
plt.xticks([])
plt.yticks([])
x=np.arange(6)
y=np.arange(6)
plt.plot(x,y)
```

```python
x=np.arange(9)
y=np.arange(10,100,10)
#画坐标系1，在2*1坐标系的的第1个坐标系中作图
plt.subplot(2,1,1)
plt.plot(x,y)
#对坐标系1画轴标签
plt.xlabel('month1')
plt.ylabel('amount1')
#对坐标系1画轴刻度
plt.tick_params(axis='x',which='major',direction='in',bottom='False')
#画坐标系2，在2*1坐标系的的第2个坐标系中作图
plt.subplot(2,1,2)
plt.bar(x,y)
#对坐标系2画轴标签
plt.xlabel('month2')
plt.ylabel('amount2')
#对坐标系2画轴刻度
plt.tick_params(axis='y',which='minor',direction='out',labelbottom='False')
```

# 坐标轴范围---plt.xlim函数

```python
plt.xlabel('month')
plt.ylabel('amount')
plt.xlim(1,10)
plt.ylim(0,1000)
x=np.arange(8)
y=np.arange(100,900,100)
plt.plot(x,y)
```

# 网格线设置---plt.grid()

```python
#网格线是相比于坐标轴更小的单位，默认关闭
plt.grid(b='True')
x=np.arange(8)
y=np.arange(100,900,100)
plt.plot(x,y)
```

```python
#可以通过修改参数axis控制打开哪个轴的网格线
#打开x轴
plt.grid(b="True",axis="x")
x=np.arange(8)
y=np.arange(100,900,100)
plt.plot(x,y)
```

```python
#还可以设置网格线的线宽-linewidth、线型-linestyle、线的颜色
#线型设置成虚线，线宽设置为1
plt.grid(b="True",linestyle='dashed',linewidth=1)
x=np.arange(8)
y=np.arange(100,900,100)
plt.plot(x,y)
```

# 设置图例---plt.legend(loc,ncol)

```python
#图例对图表起到注释作用，在绘图时先通过label参数传入值表示该图表的图例名，再通过plt.legend()方法把图例显示出来
x=np.arange(8)
y1=np.arange(100,900,100)
y2=np.arange(100,900,100)
plt.plot(x,y1,label='quantity')
plt.bar(x,y2,label='amount')
plt.legend()
```

```python
x=np.arange(8)
y1=np.arange(100,900,100)
y2=np.arange(100,900,100)
plt.plot(x,y1,label='quantity')
plt.bar(x,y2,label='amount')
plt.legend(loc='upper center')
```

```python
#可以通过ncol参数来控制显示列数
x=np.arange(8)
y1=np.arange(100,900,100)
y2=np.arange(100,900,100)
plt.plot(x,y1,label='数量')
plt.bar(x,y2,label='销量')
plt.legend(loc='upper center',ncol=2)
```

# 图表标题的设置---plt.title(label='',loc='')

```python
#参数label设置图表标题，参数loc设置显示位置，有center/left/right,还可以通过参数fontdict对标题文字的相关性质进行设置
x=np.arange(1,9,1)
y1=np.arange(100,900,100)
y2=np.arange(100,900,100)
plt.plot(x,y1,label='quantity')
plt.bar(x,y2,label='amount')
plt.legend(loc='upper left',ncol=1)
plt.title(label="1-8月销量",loc='center')
#解决mac电脑中文乱码问题
mpl.rcParams["font.sans-serif"] = ["Arial Unicode MS"]
```

# 数据标签的设置---plt.text(x,y,str,ha,va,fontsize)

```python
x=np.arange(1,9,1)
y1=np.arange(100,900,100)
y2=np.arange(100,900,100)
plt.plot(x,y1,label='quantity')
plt.text(2,200,200,ha='center',va='bottom',fontsize=10)
plt.bar(x,y2,label='amount')
plt.legend(loc='upper left',ncol=1)
plt.title(label="1-8月销量",loc='center')
#解决mac电脑中文乱码问题
mpl.rcParams["font.sans-serif"] = ["Arial Unicode MS"]
```

# 注释的设置---plt.annotate(s,xy,xytest,arrowprops)

```python
x=np.arange(1,9,1)
y1=np.arange(100,900,100)
y2=np.arange(100,900,100)
plt.plot(x,y1,label='quantity')
for a,b in zip(x,y):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=10)
plt.annotate('good',xy=(8,800),xytext=(7,800),color='red',arrowprops=dict(facecolor='red',arrowstyle='->'))
plt.bar(x,y2,label='amount')
plt.legend(loc='upper left',ncol=1)
plt.title(label="1-8月销量",loc='center')
#解决mac电脑中文乱码问题
mpl.rcParams["font.sans-serif"] = ["Arial Unicode MS"]
```

# 数据表的设置---plt.table()

```python
x=np.arange(1,9,1)
y=np.arange(100,900,100)
plt.plot(x,y,label='quantity')
for a,b in zip(x,y):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=10)
plt.annotate('good',xy=(8,800),xytext=(7,800),color='red',arrowprops=dict(facecolor='red',arrowstyle='->'))
plt.legend(loc='upper left',ncol=1)
plt.title(label="1-8月销量",loc='center')
plt.xticks([])

cellText=[[100,200,300,400,500,600,700,800]]
rows=['销量']
columns=['1','2','3','4','5','6','7','8']
plt.table(cellText=cellText,
         cellLoc='center',
         rowLabels=rows,
         rowColours=['red'],
         rowLoc='center',
         colLabels=columns,
         colColours=['red','yellow','red','yellow','red','yellow','red','yellow'],
         colLoc='left',
         loc='bottom')

#解决mac电脑中文乱码问题
mpl.rcParams["font.sans-serif"] = ["Arial Unicode MS"]
```
